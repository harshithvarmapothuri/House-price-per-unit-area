```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Real_estate[1].csv")
```

```python
df
```

```python
df.drop("No",axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={"X1 transaction date":"Date","Y house price of unit area":"Price/unit","X2 house age":"age","X3 distance to the nearest MRT station":"Nearest MRT Distance","X4 number of convenience stores":"no of convience stores","X5 latitude":"Latitude","X6 longitude":"Longitude"},inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["Date"]=df["Date"].astype(str).str[0:4]
```

```python
df
```

```python
df.rename(columns={"Date":"Year"},inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
```

```python
df.head()
```

```python
x=df.drop("Price/unit",axis=1)
```

```python
y=df["Price/unit"]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x_new=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
x_new
```

```python
import pickle
pickle.dump(scaler,open("scaler_house.pkl","wb"))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNetCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_new,y,test_size=0.25)
```

```python
model=ElasticNetCV(cv=5)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
pickle.dump(model,open("model_house.pkl","wb"))
```
